```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer;
```

```python
cancer = load_breast_cancer();
#cancer = pd.read_csv("data.csv")
x = cancer.data;
y = cancer.target;
```

## train 80% test 20%

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

## classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion = 'entropy',
                              max_depth = 3,
                              random_state = 0)
tree.fit(x_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree, feature_names = cancer.feature_names, fontsize = 6)
```

```python
from sklearn.metrics import accuracy_score;
y_pred_train = tree.predict(x_train)
print("Train set accuracy: ", accuracy_score(y_train,y_pred_train))
y_pred_test = tree.predict(x_test)
print("Test set accuracy: ", accuracy_score(y_test,y_pred_test))
```

## model run in tree depth

# 1.depth 1

```python
#entropy
tree_d1_ent = DecisionTreeClassifier(criterion = 'entropy',
                                     max_depth = 1,
                                     random_state = 0)
tree_d1_ent.fit(x_train,y_train)

y_pred_train_ent_d1 = tree_d1_ent.predict(x_train)
y_pred_test_ent_d1 = tree_d1_ent.predict(x_test)

#gini
tree_d1_gini = DecisionTreeClassifier(criterion = 'gini',
                                     max_depth = 1,
                                     random_state = 0)
tree_d1_gini.fit(x_train,y_train)

y_pred_train_gini_d1 = tree_d1_gini.predict(x_train)
y_pred_test_gini_d1 = tree_d1_gini.predict(x_test)
```
